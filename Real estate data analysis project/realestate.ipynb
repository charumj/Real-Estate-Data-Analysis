```python
import pandas as pd
import json

# Provide the full path to the JSON file
file_path = r"C:\Users\91908\Desktop\real estate data2\csvjson (1).json"

# Load the JSON file into a DataFrame
df = pd.read_json(file_path)


# Print original column names to check them
print("Original Column Names:")
print(df.columns)

# Standardize column names: strip whitespace and convert to lowercase
df.columns = df.columns.str.strip().str.lower()

# Print updated column names to verify
print("Updated Column Names:")
print(df.columns)

# Convert price ranges from crores to thousands
df['price range1'] = pd.to_numeric(df['price range1'], errors='coerce') * 10000
df['price range2'] = pd.to_numeric(df['price range2'], errors='coerce') * 10000

# Convert relevant columns to numeric, coercing errors to NaN
df['sq.ft min'] = pd.to_numeric(df['sq.ft min'], errors='coerce')
df['sq ft max'] = pd.to_numeric(df['sq ft max'], errors='coerce')

# Check for NaN values in the columns used for calculations
print("Checking for NaN values:")
print(df[['price range1', 'price range2', 'sq.ft min', 'sq ft max']].isna().sum())

# Initialize price_per_sqft as NaN
df['price_per_sqft'] = None

# Calculate price_per_sqft based on conditions
df.loc[df['price range2'].isna() | df['sq ft max'].isna(), 'price_per_sqft'] = (
    df['price range1'] / df['sq.ft min']
)

df.loc[df['price range2'].notna() & df['sq ft max'].notna(), 'price_per_sqft'] = (
    (df['price range1'] + df['price range2']) / 
    (df['sq.ft min'] + df['sq ft max'])
)

# Display the updated DataFrame with price per sqft
print(df[['city', 'name', 'price_per_sqft']])
```
